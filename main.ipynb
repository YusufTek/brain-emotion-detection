```python
from sklearn.preprocessing import LabelEncoder

emotions['label_num'] = LabelEncoder().fit_transform(emotions['label'])

print("Etiket eşleştirmesi tamamlandı.")
print(emotions[['label', 'label_num']].drop_duplicates().sort_values('label_num'))
print(f"\nHer sınıftan kaç örnek var:\n{emotions['label_num'].value_counts().sort_index()}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# Veriyi hazırla
features = emotions.drop(columns=['label', 'label_num'])
labels = emotions['label_num']

print(f"Toplam özellik sayısı: {features.shape[1]}")
print(f"Örnek sayısı: {features.shape[0]}")

# =============================================
# 1. Variance Threshold - Düşük varyans kontrolü
# =============================================
print("\n" + "="*50)
print("1. VARYANS ANALİZİ")
print("="*50)

# Her özelliğin varyansını hesapla
feature_vars = features.var().sort_values(ascending=True)

# Düşük varyanslı özellikleri göster
low_var_features = feature_vars[feature_vars < 0.01]
print(f"Düşük varyansa sahip özellik sayısı (var < 0.01): {len(low_var_features)}")

# Varyans dağılımını görselleştir
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(feature_vars, bins=50, alpha=0.7)
plt.xlabel('Varyans')
plt.ylabel('Özellik Sayısı')
plt.title('Özellik Varyanslarının Dağılımı')
plt.yscale('log')

plt.subplot(1, 2, 2)
plt.boxplot(feature_vars)
plt.ylabel('Varyans')
plt.title('Varyans Boxplot')
plt.yscale('log')

plt.tight_layout()
plt.show()
```

```python
# =============================================
# 2. Univariate Feature Selection (F-test)
# =============================================
print("\n" + "="*50)
print("2. F-TEST İLE ÖZELLİK SEÇİMİ")
print("="*50)

# F-test skorları
selector = SelectKBest(score_func=f_classif, k='all')
selector.fit(features, labels)

# Skorları DataFrame'e çevir
f_scores_df = pd.DataFrame({
    'feature': features.columns,
    'f_score': selector.scores_,
    'p_value': selector.pvalues_
})

# En iyi 45 özelliği seç
top_45_f = f_scores_df.nlargest(45, 'f_score')
print("En yüksek F-score'lu 45 özellik:")
print(top_45_f.head(45))

# F-score dağılımını göster
plt.figure(figsize=(20, 6))

plt.subplot(1, 3, 1)
plt.hist(f_scores_df['f_score'], bins=50, alpha=0.7)
plt.xlabel('F-Score')
plt.ylabel('Özellik Sayısı')
plt.title('F-Score Dağılımı')

plt.subplot(1, 3, 2)
plt.scatter(range(len(f_scores_df)), f_scores_df['f_score'].sort_values(ascending=False))
plt.xlabel('Özellik Sırası')
plt.ylabel('F-Score')
plt.title('F-Score Sıralaması')

plt.subplot(1, 3, 3)
top_45_f.head(20).plot(x='feature', y='f_score', kind='barh')
plt.title('En İyi 20 Özellik (F-Score)')
plt.xlabel('F-Score')

plt.tight_layout()
plt.show()
```

```python
# =============================================
# 3. Mutual Information
# =============================================
print("\n" + "="*50)
print("3. MUTUAL INFORMATION ANALİZİ")
print("="*50)

# Mutual Information skorları
mi_scores = mutual_info_classif(features, labels, random_state=42)
mi_scores_df = pd.DataFrame({
    'feature': features.columns,
    'mi_score': mi_scores
})

top_45_mi = mi_scores_df.nlargest(45, 'mi_score')
print("En yüksek Mutual Information'lu 45 özellik:")
print(top_45_mi.head(45))
```

```python
# =============================================
# 4. Random Forest Feature Importance
# =============================================
print("\n" + "="*50)
print("4. RANDOM FOREST ÖNEMLİLİK ANALİZİ")
print("="*50)

# Random Forest ile özellik önemliliği
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(features, labels)

rf_importance_df = pd.DataFrame({
    'feature': features.columns,
    'rf_importance': rf.feature_importances_
})

top_45_rf = rf_importance_df.nlargest(45, 'rf_importance')
print("En önemli 45 özellik (Random Forest):")
print(top_45_rf.head(45))
```

```python
# =============================================
# 5. Korelasyon Analizi
# =============================================
print("\n" + "="*50)
print("5. KORELASYON ANALİZİ")
print("="*50)

# Özellikler arası korelasyon (yüksek korelasyonlu olanları bul)
correlation_matrix = features.corr().abs()

# Üst üçgen matrisi al (kendisiyle korelasyonu hariç)
upper_triangle = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

# Yüksek korelasyonlu çiftleri bul
high_corr_pairs = []
for col in upper_triangle.columns:
    high_corr = upper_triangle[col][upper_triangle[col] > 0.95]
    for idx in high_corr.index:
        high_corr_pairs.append((col, idx, high_corr[idx]))

print(f"Yüksek korelasyonlu özellik çifti sayısı (>0.95): {len(high_corr_pairs)}")
if high_corr_pairs:
    print("İlk 45 yüksek korelasyonlu çift:")
    for i, (feat1, feat2, corr) in enumerate(high_corr_pairs[:45]):
        print(f"{feat1} - {feat2}: {corr:.3f}")
```

```python
# =============================================
# 6. Tüm Yöntemleri Birleştir
# =============================================
print("\n" + "="*50)
print("6. BİRLEŞTİRİLMİŞ SKOR ANALİZİ")
print("="*50)

# Tüm skorları normalize et ve birleştir
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Skorları normalize et
normalized_f = scaler.fit_transform(f_scores_df[['f_score']]).flatten()
normalized_mi = scaler.fit_transform(mi_scores_df[['mi_score']]).flatten()
normalized_rf = scaler.fit_transform(rf_importance_df[['rf_importance']]).flatten()

# Birleştirilmiş skor
combined_scores_df = pd.DataFrame({
    'feature': features.columns,
    'f_score_norm': normalized_f,
    'mi_score_norm': normalized_mi,
    'rf_score_norm': normalized_rf,
    'combined_score': (normalized_f + normalized_mi + normalized_rf) / 3
})

# En iyi özellikleri seç
top_features_combined = combined_scores_df.nlargest(50, 'combined_score')
print("En iyi 50 özellik (birleştirilmiş skor):")
print(top_features_combined.head(50))
```

```python
# =============================================
# 7. En İyi Özellikleri Görselleştir
# =============================================
print("\n" + "="*50)
print("7. EN İYİ ÖZELLİKLERİN GÖRSELLEŞTİRİLMESİ")
print("="*50)

# En iyi 45 özellik için boxplot
best_45_features = top_features_combined.head(45)['feature'].tolist()

fig, axes = plt.subplots(9, 5, figsize=(20, 36))
axes = axes.flatten()

for i, feature in enumerate(best_45_features):
    emotions.boxplot(column=feature, by='label_num', ax=axes[i])
    axes[i].set_title(f'{feature}\n(Score: {top_features_combined.iloc[i]["combined_score"]:.3f})', 
                     fontsize=10)
    axes[i].set_xlabel('Emotion')

plt.suptitle('En İyi 16 Özellik - Boxplot Analizi', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# =============================================
# 8. Özellik Türü Analizi
# =============================================
print("\n" + "="*50)
print("8. ÖZELLİK TÜRÜ ANALİZİ")
print("="*50)

# Özellik türlerini analiz et
feature_types = {}
for feature in top_features_combined.head(45)['feature']:
    if feature.startswith('mean_'):
        feature_type = 'mean'
    elif feature.startswith('fft_'):
        feature_type = 'fft'
    else:
        feature_type = 'other'
    
    if feature_type not in feature_types:
        feature_types[feature_type] = 0
    feature_types[feature_type] += 1

print("En iyi 45 özellik türü dağılımı:")
for ftype, count in feature_types.items():
    print(f"{ftype}: {count} özellik")

# Önerilen özellik listesi
print("\n" + "="*60)
print("ÖNERİLEN ÖZELLİK LİSTESİ (Machine Learning için)")
print("="*60)
recommended_features = top_features_combined.head(45)['feature'].tolist()
print("En iyi 45 özellik:")
for i, feature in enumerate(recommended_features, 1):
    score = top_features_combined.iloc[i-1]['combined_score']
    print(f"{i:2d}. {feature:<15} (skor: {score:.4f})")

# Son özet
print(f"\nÖNERİ: Bu {len(recommended_features)} özelliği kullanarak model eğitimi yapabilirsiniz.")
print("Bu özellikler farklı yöntemlerle en yüksek ayırt edicilik skorunu almışlardır.")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Farklı özellik sayılarını test et
results = {}
for k in [10, 20, 30, 36, 45, 60]:
    selected_features = top_features_combined.head(k)['feature'].tolist()
    X_selected = features[selected_features]
    
    rf = RandomForestClassifier(random_state=42)
    scores = cross_val_score(rf, X_selected, labels, cv=5)
    results[k] = scores.mean()
    print(f"{k} özellik: {scores.mean():.4f} ± {scores.std():.4f}")
```

```python
optimal_features = top_features_combined.head(45)['feature'].tolist()
X_final, y = features[optimal_features].values, labels.values

print(f"\nOptimal özellik sayısı: {len(optimal_features)}")
print(f"Optimal özellikler: {optimal_features}")
```

```python
from sklearn.model_selection import train_test_split

X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42, stratify=y)

print(f"Train set boyutu: {X_final_train.shape[0]}")
print(f"Test set boyutu: {X_final_test.shape[0]}")
```

```python
from sklearn.linear_model import LogisticRegression

reg = 0.01

multi_model = LogisticRegression(C=1/reg, solver='saga', max_iter=10000, random_state=42)

multi_model.fit(X_final_train, y_train)
```

```python
emo_pred = multi_model.predict(X_final_test)

print(f"Test setindeki tahminler: {emo_pred[:10]}")
print(f"Gerçek etiketler        : {y_test[:10]}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Sınıflandırma raporu:")
print(classification_report(y_test, emo_pred))
print("Karmaşıklık matrisi:")
cm = confusion_matrix(y_test, emo_pred)
print(cm)
```

```python
from sklearn.metrics import recall_score, precision_score, accuracy_score

print(f"Accuracy: {accuracy_score(y_test, emo_pred)}")
print(f"Precision: {precision_score(y_test, emo_pred, average='macro')}")
print(f"Recall: {recall_score(y_test, emo_pred, average='macro')}")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

feature_columns = list(range(45))

feature_transformer = Pipeline(
    steps=[
        ('scaler', StandardScaler())
    ])

preprocessor = ColumnTransformer(
    transformers=[
        ('preprocess', feature_transformer, feature_columns)
    ])

pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('regressor', SVC(probability=True))
    ])

multi_model = pipeline.fit(X_final_train, y_train)
print(multi_model)
```

```python
from sklearn.metrics import f1_score, roc_auc_score

emo_pred = multi_model.predict(X_final_test)
emo_prob = multi_model.predict_proba(X_final_test)

print(f"Accuracy: {accuracy_score(y_test, emo_pred)}")
print(f"Recall: {recall_score(y_test, emo_pred, average='macro')}")
print(f"Precision: {precision_score(y_test, emo_pred, average='macro')}")
print(f"F1 Score: {f1_score(y_test, emo_pred, average='macro')}")
print(f"ROC AUC: {roc_auc_score(y_test, emo_prob, multi_class='ovr')}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_pipeline = Pipeline(
    steps=[
        # İlk adım: Veri standardizasyonu
        # Random Forest normalde scaling gerektirmez (tree-based method)
        # Ancak diğer modellerle karşılaştırma tutarlılığı için eklendi
        ('scaler', StandardScaler()),
        ('classifier', RandomForestClassifier(
            # n_estimators: Ormandaki karar ağacı sayısı
            # Daha fazla ağaç = daha iyi performans ancak daha yavaş
            # 100: Performans/hız dengesinde optimal değer
            n_estimators=100,
            # random_state: Reproducible sonuçlar için sabit seed
            # Aynı kodun her çalıştırılmasında aynı sonuçları verir
            random_state=42,
            # max_depth: Her ağacın maksimum derinliği
            # 20: Overfitting'i önlerken yeterli karmaşıklığa izin verir
            # None olsaydı: Ağaçlar tam büyüyebilirdi (overfitting riski)
            max_depth=20,
            # min_samples_split: Bir node'un split olması için gereken min örnek sayısı
            # 5: Çok küçük split'leri önler, generalization'ı artırır
            # Değer ne kadar büyük = o kadar konservatif model
            min_samples_split=5,
            # min_samples_leaf: Her leaf node'da bulunması gereken min örnek sayısı
            # Default: 1, burada belirtilmemiş ama 2-5 arası iyi değerler
            # min_samples_leaf=2,  # İsterseniz ekleyebilirsiniz
            
            # bootstrap: Her ağaç için farklı alt örnek kümesi kullan
            # True (default): Bagging tekniği, variance'ı azaltır
            # bootstrap=True,
            
            # max_features: Her split'te göz önünde bulundurulan özellik sayısı
            # 'sqrt': sqrt(n_features) kadar özellik (genelde iyi sonuç verir)
            # max_features='sqrt',  # Opsiyonel, default genelde iyidir
            
            # n_jobs: Paralel işlem sayısı
            # -1: Tüm CPU core'larını kullan (hızlandırır)
            n_jobs=-1

            # class_weight: Dengesiz veri setleri için sınıf ağırlıkları
            # 'balanced': Minority class'lara daha fazla ağırlık verir
            # Bizim veri setimiz dengeli olduğu için None (default)
            # class_weight='balanced'  # Gerekirse ekleyin
        ))
    ])

rf_model = rf_pipeline.fit(X_final_train, y_train)
rf_pred = rf_model.predict(X_final_test)
rf_prob = rf_model.predict_proba(X_final_test)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, rf_pred)}")
print(f"Recall Score: {recall_score(y_test, rf_pred, average='macro')}")
print(f"Precision Score: {precision_score(y_test, rf_pred, average='macro')}")
print(f"F1 Score: {f1_score(y_test, rf_pred, average='macro')}")
print(f"ROC AUC Score: {roc_auc_score(y_test, rf_prob, multi_class='ovr')}")
```

```python
import pandas as pd
import numpy as np
import os

# Orijinal emotions.csv dosyasını yükle
dir_file = os.getcwd()
emotions = pd.read_csv(os.path.join(dir_file, 'emotions.csv'))

# LabelEncoder'ı yeniden oluştur (aynı eşleştirmeyi yapmak için)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
emotions['label_num'] = le.fit_transform(emotions['label'])

# Raw değerlerin hangi duygulara karşılık geldiğini göster
print("=== RAW DEĞERLERİN DUYGU EŞLEŞTİRMESİ ===")
label_mapping = emotions[['label', 'label_num']].drop_duplicates().sort_values('label_num')
print(label_mapping)

print("\n=== DETAYLI EŞLEŞTİRME ===")
for idx, row in label_mapping.iterrows():
    print(f"Raw {row['label_num']} = {row['label']}")

# Ters eşleştirme için dictionary oluştur
raw_to_emotion = dict(zip(label_mapping['label_num'], label_mapping['label']))
emotion_to_raw = dict(zip(label_mapping['label'], label_mapping['label_num']))

print("\n=== SORGULAMA ÖRNEKLERİ ===")
test_values = [0, 1, 2, 3, 4, 5, 6]
for val in test_values:
    if val in raw_to_emotion:
        print(f"Raw {val} = {raw_to_emotion[val]}")
    else:
        print(f"Raw {val} = Bilinmeyen")

print("\n=== TERSİNE SORGULAMA ===")
for emotion in ['NEGATIVE', 'NEUTRAL', 'POSITIVE']:
    if emotion in emotion_to_raw:
        print(f"{emotion} = Raw {emotion_to_raw[emotion]}")
    else:
        print(f"{emotion} = Bu etiket veri setinde yok")
```
